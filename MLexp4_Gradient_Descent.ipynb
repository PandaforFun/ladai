```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return (x + 3)**2

def gradient(x):
    return 2 * (x + 3)

# Gradient Descent Algorithm
def gradient_descent(learning_rate=0.1, max_steps=100, tolerance=1e-6, x_init=2):
    x = x_init
    steps_taken = 0
    x_history = [x] 

    for _ in range(max_steps):
        grad = gradient(x)
        new_x = x - learning_rate * grad
        x_history.append(new_x)

        if abs(new_x - x) < tolerance:
            break

        x = new_x
        steps_taken += 1

    return x, x_history, steps_taken

# Parameters
learning_rate = 0.1
x_init = 2
max_steps = 100
tolerance = 1e-6

# Running gradient descent
x_min, x_history, steps_taken = gradient_descent(learning_rate, max_steps, tolerance, x_init)

# Plotting the function and gradient descent path
x_vals = np.linspace(-10, 5, 100)
y_vals = func(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='y = (x + 3)^2')
plt.scatter(x_history, [func(x) for x in x_history], color='red', marker='x', label='Gradient Descent Steps')
plt.title('Gradient Descent Visualization')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print(f"Local minimum occurs at x = {x_min}, found in {steps_taken} steps.")
```
